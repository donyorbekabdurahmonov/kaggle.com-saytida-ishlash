```python
# Importamos las librerias necesarias
import pandas as pd

# Cargamos nuestro dataframe
df = pd.read_csv('/kaggle/input/viajeros/sevilla_viajeros_por_franja_csv.csv', sep=';')

# Mostramos el dataframe cargado
df
```

## Limpiar outliers

```python
# Histograma (cuantas veces se repite un valor)
df['VIAJEROS_SUBIDOS'].hist()
```

```python
# Eliminamos todas las filas que en la columna 'VIAJEROS_SUBIDOS' superen los 100
df.drop(df[(df['VIAJEROS_SUBIDOS'] > 100)].index, inplace=True)

# Restablecemos el índice
df.reset_index(inplace=True)

# Volvemos a mostrar el histograma
df['VIAJEROS_SUBIDOS'].hist()
```

Ahora hay menos outliers

## Limpiar nulos

```python
# Comprobar las filas que tienen algún valor nulo
df[df.isnull().any(axis=1)]
```

Ya que no hay ningún valor nulo, no hacemos nada
